```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from scipy.sparse import data
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics.cluster import rand_score, fowlkes_mallows_score, adjusted_rand_score
from sklearn.metrics import jaccard_score, precision_score
from yellowbrick.cluster import KElbowVisualizer
```

```python
from google.colab import files
uploaded = files.upload()
data1 = pd.read_csv('Data1.csv')
data3 = pd.read_csv('Data3.csv')
data4 = pd.read_csv('Data4.csv')
data5 = pd.read_csv('Data5.csv')
data6 = pd.read_csv('Data6.csv')
data7 = pd.read_csv('Data7.csv')
data8 = pd.read_csv('Data8.csv')
data2a = pd.read_csv('Data2.csv')
data2 = data2a.rename(columns={"X": "X1", "Y": "X2", "C": "X3"}, inplace= True)
data2 = data2a.drop(data2a.columns[0],axis=1)
```

##Data 1

```python
# TASK 1.1 Use K-means and hierarchical clustering methods to generate cluster
print("\n1.1A KMEANS CLUSTERING OF DATA 1 \n")
model1 = KMeans()
visualizer = KElbowVisualizer(model1, k =(1,15)).fit(data1.iloc[:,1:4])
visualizer.show()
print('No of clusters using KMeans Algorithm', visualizer.elbow_value_, '\n')
k_means_optimum1 = KMeans(n_clusters = 6, init = 'k-means++',  random_state=0)
No_of_clusters_KMeans1 = k_means_optimum1.fit_predict(data1.iloc[:,1:4])
print(No_of_clusters_KMeans1)

#Task 1.1B hierarchical clustering methods to generate cluster
print("\n1.1B HIERARCHICAL CLUSTERING OF DATA 1\n")
plt.title("Dendrograms")  
shc.dendrogram(shc.linkage(data1.iloc[:,1:4], method='ward'))
plt.axhline(y=15, color='r', linestyle='--')
print('No of clusters using Hierarchical algorithm = 7 \n')
cluster = AgglomerativeClustering(n_clusters=7, affinity='euclidean', linkage='ward')  
No_of_clusters_Hier1 = cluster.fit_predict(data1.iloc[:,1:4])
print(No_of_clusters_Hier1)
```

```python
# Task 1.2 Evaluate the performance of the clustering algorithm using external validation metrics
KM1 = No_of_clusters_KMeans1.tolist()
HC1 = No_of_clusters_Hier1.tolist()
print('\n 1.2 PERFORMANCE METRICS OF KMEANS CLUSTERING OF DATA 1\n')
print('RAND SCORE', rand_score(KM1, data1['Class']))
print('FOWLKES MALLOWS SCORE', fowlkes_mallows_score(KM1, data1['Class']))
print('PRECISION SCORE', precision_score(KM1, data1['Class'], average = 'weighted'))
print('\n 1.2 PERFORMANCE METRICS OF HIERARCHICAL CLUSTERING OF DATA 1\n')
print('RAND SCORE', rand_score(data1['Class'], HC1))
print('FOWLKES MALLOWS SCORE', fowlkes_mallows_score(HC1, data1['Class']))
print('PRECISION SCORE', precision_score(HC1, data1['Class'], average = 'weighted'))

# Task 1.3 Plot (2D or 3D) the data points for each dataset and color them according to the original class
print("\n 1.3 VISUALISING THE DATA POINTS USING ORIGINAL CLUSTERS OF DATA 1\n")
plt.figure(figsize=(10,7))
groups1 = data1.groupby("Class")
ax = plt.axes(projection="3d")
for Class, group in groups1:
  ax.plot(group["X1"], group["X2"], group["X3"], marker=".", linestyle="", label= Class)
ax.legend(loc = 'upper left')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
plt.show()

# Task 1.4 Plot (2D or 3D) the data points for each dataset and color them according to the class allocated by the clustering algorithm
print("\n 1.4A VISUALISING THE DATA POINTS USING KMEANS CLUSTERS OF DATA 1\n")
plt.figure(figsize=(10,7))
data1['Cluster_KM'] = No_of_clusters_KMeans1
kplot = plt.axes(projection='3d')
kplot.scatter3D(data1[data1.Cluster_KM==0].X1, data1[data1.Cluster_KM==0].X2, data1[data1.Cluster_KM==0].X3, c='red', label = 'Cluster 1')
kplot.scatter3D(data1[data1.Cluster_KM==1].X1, data1[data1.Cluster_KM==1].X2, data1[data1.Cluster_KM==1].X3,c ='green', label = 'Cluster 2')
kplot.scatter3D(data1[data1.Cluster_KM==2].X1, data1[data1.Cluster_KM==2].X2, data1[data1.Cluster_KM==2].X3,c ='blue', label = 'Cluster 3')
kplot.scatter3D(data1[data1.Cluster_KM==3].X1, data1[data1.Cluster_KM==3].X2, data1[data1.Cluster_KM==3].X3,c ='cyan', label = 'Cluster 4')
kplot.scatter3D(data1[data1.Cluster_KM==4].X1, data1[data1.Cluster_KM==4].X2, data1[data1.Cluster_KM==4].X3,c ='black', label = 'Cluster 5')
kplot.scatter3D(data1[data1.Cluster_KM==5].X1, data1[data1.Cluster_KM==5].X2, data1[data1.Cluster_KM==5].X3,c ='yellow', label = 'Cluster 6')
plt.legend(loc = 'upper left')
plt.title("Kmeans")
plt.show()

print("\n 1.4B VISUALISING THE DATA POINTS USING HIERARCHICAL CLUSTERS OF DATA 1\n")
plt.figure(figsize=(10,7))
data1['Cluster_HC'] = No_of_clusters_Hier1
kplot = plt.axes(projection='3d')
kplot.scatter3D(data1[data1.Cluster_HC==0].X1, data1[data1.Cluster_HC==0].X2, data1[data1.Cluster_HC==0].X3,c='red', label = 'Cluster 1')
kplot.scatter3D(data1[data1.Cluster_HC==1].X1, data1[data1.Cluster_HC==1].X2, data1[data1.Cluster_HC==1].X3,c ='green', label = 'Cluster 2')
kplot.scatter3D(data1[data1.Cluster_HC==2].X1, data1[data1.Cluster_HC==2].X2, data1[data1.Cluster_HC==2].X3,c ='blue', label = 'Cluster 3')
kplot.scatter3D(data1[data1.Cluster_HC==3].X1, data1[data1.Cluster_HC==3].X2, data1[data1.Cluster_HC==3].X3,c ='cyan', label = 'Cluster 4')
kplot.scatter3D(data1[data1.Cluster_HC==4].X1, data1[data1.Cluster_HC==4].X2, data1[data1.Cluster_HC==4].X3,c ='black', label = 'Cluster 5')
kplot.scatter3D(data1[data1.Cluster_HC==5].X1, data1[data1.Cluster_HC==5].X2, data1[data1.Cluster_HC==5].X3,c ='yellow', label = 'Cluster 6')
kplot.scatter3D(data1[data1.Cluster_HC==6].X1, data1[data1.Cluster_HC==6].X2, data1[data1.Cluster_HC==6].X3,c ='violet', label = 'Cluster 7')
plt.legend(loc = 'upper left')
plt.title("HIERARCHICAL")
plt.show()
```

## Conclusion


For all the data, we have developed clusters using two most commonly used clustering algorithms (KMeans and Hierarchical) and evaluated by different evaluation metrics to generate the accuracy of the predictions generated. Metrics like rand and precision will lead to poor scores as the model assigned labels are completely different from the original labels in value. Fowlkes Mallows score checks the similarity of the clusters and even if the labels are different from each other. Hence, it develops a high score if the datapoints are clustered as same.
